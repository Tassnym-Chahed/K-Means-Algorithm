```python
import pandas as pd
from matplotlib import pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
# Dataset
df = pd.read_csv('/Users/tassnymechchahed/Desktop/force2020_data_unsupervised_learning.csv', index_col = 'DEPTH_MD')
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
# Transform Data - Standardize it
df.describe()
scaler = StandardScaler()   
df[['RHOB_T', 'NPHI_T', 'GR_T', 'PEF_T', 'DTC_T']] = scaler.fit_transform(df[['RHOB', 'NPHI', 'GR', 'PEF', 'DTC']])
df
```

```python
# Choose right number of clusters
def optimise_k_means(data, max_k):
    means=[]
    inertias=[]

    for k in range(1, max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)

        means.append(k)
        inertias.append(kmeans.inertia_)

    # Generate the elbow plot
    fig,ax = plt.subplots(figsize=(10,5))
    plt.plot(means, inertias, 'o-')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()


```

```python
optimise_k_means(df[['RHOB_T', 'NPHI_T']], 10)
```

```python
# Applying K Means Clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(df[['NPHI_T', 'RHOB_T']])
```

```python
df['kmeans_3'] = kmeans.labels_
df
```

```python
# Ploy Results
plt.scatter(x=df['NPHI'], y=df['RHOB'], c=df['kmeans_3'])
plt.xlim(-0.1,1)
plt.ylim(3,1.5)
plt.show()
```
